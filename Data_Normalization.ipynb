```python
import numpy as np
from numpy import nan
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
wa1 = pd.read_csv('data/Allchr_030913_A33_W_ls6.txt', sep="\t")
wa2 = pd.read_csv('data/Allchr_062216_A33_W_ls6.txt', sep="\t")
ra1 = pd.read_csv('data/Allchr_030913_A34_r_ls6.txt', sep="\t")
ra2 = pd.read_csv('data/Allchr_062216_A34_r_ls6.txt', sep="\t")

ww1 = pd.read_csv('data/Allchr_051313_A139_W_ls6.txt', sep="\t")
ww2 = pd.read_csv('data/Allchr_062216_A139_W_ls6.txt', sep="\t")
rw1 = pd.read_csv('data/Allchr_051313_A4095_r_ls6.txt', sep="\t")
rw2 = pd.read_csv('data/Allchr_062216_A4095_r_ls6.txt', sep="\t")

wa1.head()
```

```python
size = [2.30218, 8.13184, 3.1662, 15.31933, 5.76874, 2.70161, 10.9094, 5.62643, 4.39888, 7.45751, 6.66816, 10.78177, 9.24431, 7.84333, 10.91291, 9.48066]

Figures = []

i = 2
while i < 18:
    globals()['Figure{}'.format(i-1)] = plt.figure(figsize=(size[i-2], 3))
    plt.plot(ww1['Coordinates (kb)'], ww1.iloc[:, i],'r-')
    plt.plot(rw1['Coordinates (kb)'], rw1.iloc[:, i],'b--')
    plt.title("chromsome{}".format(i-1))
    Figures.append(globals()['Figure{}'.format(i-1)])
    i += 1
```

```python
wa1_305 = wa1.iloc[79:240]['Chr3_ssDNA_loess'] #20kb to 60kb
wa2_305 = wa2.iloc[79:240]['Chr3_ssDNA_loess']
ra1_305 = ra1.iloc[79:240]['Chr3_ssDNA_loess'] #20kb to 60kb
ra2_305 = ra2.iloc[79:240]['Chr3_ssDNA_loess']

ww1_305 = ww1.iloc[79:240]['Chr3_ssDNA_loess'] #20kb to 60kb
ww2_305 = ww2.iloc[79:240]['Chr3_ssDNA_loess']
rw1_305 = rw1.iloc[79:240]['Chr3_ssDNA_loess'] #20kb to 60kb
rw2_305 = rw2.iloc[79:240]['Chr3_ssDNA_loess']

wa1_305.head()
```

```python
# Max and Min at ARS305 in A364a experiment 1
Max_wa1 = wa1_305.max()
Max_ra1 = ra1_305.max()
Max_ra1_new = Max_wa1*1.408074072
Min_ra1 = ra1_305.min()
print('Max_wt_A364a_exp1: ', Max_wa1)
print('Max_rad53_A364a_exp1: ', Max_ra1)
print('Max_rad53_new_A364a_exp1: ', Max_ra1_new)
print('Min_rad53_A364a_exp1: ', Min_ra1)
print("###############################")

# Max and Min at ARS305 in A364a experiment 2
Max_wa2 = wa2_305.max()
Max_ra2 = ra2_305.max()
Max_ra2_new = Max_wa2*1.408074072
Min_ra2 = ra2_305.min()
print('Max_wt_A364a_exp2: ', Max_wa2)
print('Max_rad53_A364a_exp2: ', Max_ra2)
print('Max_rad53_new_A364a_exp2: ', Max_ra2_new)
print('Min_rad53_A364a_exp2: ', Min_ra2)
print("###############################")

# Max and Min at ARS305 in W303 experiment 1
Max_ww1 = ww1_305.max()
Max_rw1 = rw1_305.max()
Max_rw1_new = Max_ww1*0.992569984
Min_rw1 = rw1_305.min()
print('Max_wt_W303_exp1: ', Max_ww1)
print('Max_rad53_W303_exp1: ', Max_rw1)
print('Max_rad53_new_W303_exp1: ', Max_rw1_new)
print('Min_rad53_W303_exp1: ', Min_rw1)
print("###############################")

# Max and Min at ARS305 in W303 experiment 2
Max_ww2 = ww2_305.max()
Max_rw2 = rw2_305.max()
Max_rw2_new = Max_ww2*0.992569984
Min_rw2 = rw2_305.min()
print('Max_wt_W303_exp2: ', Max_ww2)
print('Max_rad53_W303_exp2: ', Max_rw2)
print('Max_rad53_new_W303_exp2: ', Max_rw2_new)
print('Min_rad53_W303_exp2: ', Min_rw2)
print("###############################")
```

```python
All_chr = wa1[['Coordinates (kb)', 'Coordinates (bp)']]
Cols = ['Coordinates (kb)', 
        'Coordinates (bp)', 
        'Chr1_ssDNA_new', 
        'Chr2_ssDNA_new', 
        'Chr3_ssDNA_new', 
        'Chr4_ssDNA_new', 
        'Chr5_ssDNA_new',
        'Chr6_ssDNA_new',
        'Chr7_ssDNA_new', 
        'Chr8_ssDNA_new', 
        'Chr9_ssDNA_new', 
        'Chr10_ssDNA_new', 
        'Chr11_ssDNA_new',
        'Chr12_ssDNA_new',
        'Chr13_ssDNA_new', 
        'Chr14_ssDNA_new', 
        'Chr15_ssDNA_new',
        'Chr16_ssDNA_new',
       ]
```

```python
# Nomralize A364a experiment 1
ra1_n = ra1
ra1_n = ra1_n.drop('Coordinates (kb)', 1)
ra1_n = ra1_n.drop('Coordinates (bp)', 1)
ra1_n = ((ra1_n - Min_ra1)/(Max_ra1 - Min_ra1))*(Max_ra1_new - Min_ra1) + Min_ra1
result1 = pd.concat([All_chr, ra1_n], axis=1)
result1.columns =  Cols
wa1.columns = Cols
result1.to_csv('Normalized_data/A364a_rad53_exp1_new.csv', index=False)
wa1.to_csv('Normalized_data/A364a_WT_exp1_new.csv', index=False)

# Nomralize A364a experiment 2
ra2_n = ra2
ra2_n = ra2_n.drop('Coordinates (kb)', 1)
ra2_n = ra2_n.drop('Coordinates (bp)', 1)
ra2_n = ((ra2_n - Min_ra2)/(Max_ra2 - Min_ra2))*(Max_ra2_new - Min_ra2) + Min_ra2
result2 = pd.concat([All_chr, ra2_n], axis=1)
result2.columns =  Cols
wa2.columns = Cols
result2.to_csv('data/A364a_rad53_exp2_new.csv', index=False)
wa2.to_csv('data/A364a_WT_exp2_new.csv', index=False)

# Nomralize W303 experiment 1
rw1_n = rw1
rw1_n = rw1_n.drop('Coordinates (kb)', 1)
rw1_n = rw1_n.drop('Coordinates (bp)', 1)
rw1_n = ((rw1_n - Min_rw1)/(Max_rw1 - Min_rw1))*(Max_rw1_new - Min_rw1) + Min_rw1
result3 = pd.concat([All_chr, rw1_n], axis=1)
result3.columns =  Cols
ww1.columns = Cols
result3.to_csv('data/W303_rad53_exp1_new.csv', index=False)
ww1.to_csv('data/W303_WT_exp1_new.csv', index=False)

# Nomralize W303 experiment 2
rw2_n = rw2
rw2_n = rw2_n.drop('Coordinates (kb)', 1)
rw2_n = rw2_n.drop('Coordinates (bp)', 1)
rw2_n = ((rw2_n - Min_rw2)/(Max_rw2 - Min_rw2))*(Max_rw2_new - Min_rw2) + Min_rw2
result4 = pd.concat([All_chr, rw2_n], axis=1)
result4.columns =  Cols
ww2.columns = Cols
result4.to_csv('data/W303_rad53_exp2_new.csv', index=False)
ww2.to_csv('data/W303_WT_exp2_new.csv', index=False)
```
